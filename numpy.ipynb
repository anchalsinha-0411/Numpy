Q5. Swap columns 1 and 2 in the array.

```python
import numpy as np

# Assuming 'test' is the array you want to modify based on previous cells.
# If you have a different array, please replace 'test' with the correct variable name.

# Create a sample array if 'test' is not defined in the current session
try:
    test
except NameError:
    test = np.arange(9).reshape(3,3)
    print("Created a sample array 'test':")
    print(test)
    print("-" * 20)


# Swap columns 1 and 2 (remembering that indexing starts at 0)
# The slicing [:, [1, 0, 2]] selects all rows (:)
# and specifies the order of columns: column 1, then column 0, then column 2
test[:, [0, 1]] = test[:, [1, 0]]
print("Array with columns 1 and 2 swapped:")
print(test)
```

Q6.swap rows 1 and 2 in the array


```python
import numpy as np

# Assuming 'test' is the array you want to modify based on previous cells.
# If you have a different array, please replace 'test' with the correct variable name.

# Create a sample array if 'test' is not defined in the current session
try:
    test
except NameError:
    test = np.arange(9).reshape(3,3)
    print("Created a sample array 'test':")
    print(test)
    print("-" * 20)


# Swap rows 1 and 2 (remembering that indexing starts at 0)
# The slicing [[1, 0, 2], :] selects the rows in the order of row 1, then row 0, then row 2
# and selects all columns (:)
test[[0, 1], :] = test[[1, 0], :]
print("Array with rows 1 and 2 swapped:")
print(test)
```
